### Feature Name                                        Meaning                                                                                 

- tGravityAcc-mean()-X             - Mean of gravity acceleration signal along X-axis (time domain)                               
- tGravityAcc-min()-X              - Minimum value of gravity acceleration signal along X-axis                                    
- angle(X,gravityMean)             - Angle between X-axis and the gravity mean vector                                             
- tGravityAcc-mean()-Y             - Mean of gravity acceleration signal along Y-axis                                             
- angle(Y,gravityMean)             - Angle between Y-axis and the gravity mean vector                                             
- tGravityAcc-max()-Y              - Maximum of gravity acceleration signal along Y-axis                                          
- tGravityAcc-min()-Y              - Minimum of gravity acceleration signal along Y-axis                                         
- tGravityAcc-energy()-X           - Energy (sum of squares) of gravity acceleration signal along X-axis                          
- tGravityAcc-max()-X              - Maximum of gravity acceleration signal along X-axis                                          
- tGravityAcc-energy()-Y           - Energy of gravity acceleration signal along Y-axis                                           
- tGravityAccMag-std()             - Standard deviation of magnitude of gravity acceleration vector                               
- fBodyAccJerk-bandsEnergy()-1,16  - Energy in the 1â€“16 Hz band of the Jerk signal from body acceleration in the frequency domain 
- tGravityAcc-arCoeff()-Z,2        - 2nd autoregression coefficient of gravity acceleration signal on Z-axis                      
- tGravityAcc-min()-Z              - Minimum of gravity acceleration signal along Z-axis                                          
- tGravityAcc-mean()-Z             - Mean of gravity acceleration signal along Z-axis

```python
train_df = pd.read_csv(r"C:\Users\admin\Downloads\HAR_Train_.csv")
feature_df = train_df.drop(['Subject', 'Activity'], axis=1)

# IQR Outlier Handling
Q1 = feature_df.quantile(0.25)
Q3 = feature_df.quantile(0.75)
IQR = Q3 - Q1

feature_df = feature_df.clip(lower=(Q1 - 1.5 * IQR), upper=(Q3 + 1.5 * IQR), axis=1)

train_df.update(feature_df)
```

```python
X_train = train_df.drop(columns=['Subject', 'Activity'])
y_train = train_df['Activity']

X_test = test_df.drop(columns=['Subject', 'Activity'])
y_test = test_df['Activity']
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "XGBoost": XGBClassifier()
}
```

```python
for name, model in models.items():
    print(f"\n==== {name} ====")

    # --- XGBoost needs labels starting at 0 -----------------
    if isinstance(model, XGBClassifier) or name.lower().startswith("xgboost"):
        model.fit(X_train_scaled, y_train - 1)          # shift labels down
        y_train_pred = model.predict(X_train_scaled) + 1  # shift predictions back
        y_test_pred  = model.predict(X_test_scaled)  + 1
    # --------------------------------------------------------
    else:
        model.fit(X_train_scaled, y_train)
        y_train_pred = model.predict(X_train_scaled)
        y_test_pred  = model.predict(X_test_scaled)

    # ----- common evaluation -----
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc  = accuracy_score(y_test,  y_test_pred)

    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy : {test_acc:.4f}")
    print("\nClassification Report (Test):")
    print(classification_report(y_test, y_test_pred))
    # print(confusion_matrix(y_test, y_test_pred))
```

```python
param_grids = {
    "Logistic Regression": {
        'C': np.logspace(-3, 3, 10),
        'solver': ['liblinear', 'saga'],
        'penalty': ['l1', 'l2'],
    },
    "Decision Tree": {
        'max_depth': [3, 5, 10, 20, None],
        'min_samples_split': [2, 5, 10],
        'criterion': ['gini', 'entropy'],
    },
    "Random Forest": {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5],
    },
    "AdaBoost": {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1.0],
    },
    "Gradient Boosting": {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5, 10],
    },
    "XGBoost": {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10],
        'subsample': [0.6, 0.8, 1.0],
    }
}
```

```python

best_models = {}

for name, model in models.items():
    print(f"\n=== Hyperparameter Tuning: {name} ===")

    # Adjust y for XGBoost (classes must start at 0)
    y_train_mod = y_train - 1 if name == 'XGBoost' else y_train

    search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grids[name],
        n_iter=10,
        cv=3,
        scoring='accuracy',
        random_state=42,
        n_jobs=-1,
        verbose=1
    )

    search.fit(X_train_scaled, y_train_mod)

    best_model = search.best_estimator_

    # Predict & correct class shift (for XGBoost)
    if name == 'XGBoost':
        y_train_pred = best_model.predict(X_train_scaled) + 1
        y_test_pred = best_model.predict(X_test_scaled) + 1
    else:
        y_train_pred = best_model.predict(X_train_scaled)
        y_test_pred = best_model.predict(X_test_scaled)

    # Store best model
    best_models[name] = best_model

    # Evaluate
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    print(f"Best Parameters: {search.best_params_}")
    print(f"Train Accuracy : {train_acc:.4f}")
    print(f"Test Accuracy  : {test_acc:.4f}")
```
